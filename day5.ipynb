```julia
function readfile(filename::String)::String
    f = open(filename, "r")
    s = read(f, String)
    close(f)
    return strip(s)
end

readfile("day5.example")
```

```julia
split_parts(s::String)::Array{String} = split(s, "\n\n")

split_parts(readfile("day5.example"))
```

```julia
split_lines(s::String)::Array{String} = split(s, '\n')

split_lines.(split_parts(readfile("day5.example")))
```

```julia
function parse_range(s::String)::UnitRange{Int64}
    b,e = split(s, '-')
    return parse(Int64, b) : parse(Int64, e)
end

parse_id(s::String)::Int64 = parse(Int64, s)

tmp = map((xs, f) -> f.(xs), split_lines.(split_parts(readfile("day5.example"))), (parse_range, parse_id))
```

```julia
function check_id(id::Int64, ranges::Array{UnitRange{Int64}})::Bool
    @simd for range in ranges
        if id in range
            return true
        end
    end
    return false
end

function solve_part1(ranges::Array{UnitRange{Int64}}, ids::Array{Int64})::Int64
    return sum([check_id(id, ranges) for id in ids])
end

solve_part1(map((xs, f) -> f.(xs), split_lines.(split_parts(readfile("day5.example"))), (parse_range, parse_id))...)
```

```julia
@time solve_part1(map((xs, f) -> f.(xs), split_lines.(split_parts(readfile("day5.input"))), (parse_range, parse_id))...)
```

```julia
function solve_part2(ranges::Array{UnitRange{Int64}})::Int64
    rmin = min(first.(ranges)...)
    rmax = max(last.(ranges)...)
    println(rmin, ' ', rmax)
    #return sum(length.(ranges)) # NOTE: FIXME: this is just for test
    n = 0
    for id in rmin:rmax
        n += check_id(id, ranges)
    end
    return n
end

solve_part2(parse_range.(split_lines(split_parts(readfile("day5.example"))[1])))
```
