```julia
function readfile(filename::String)::String
    f = open(filename, "r")
    s = read(f, String)
    close(f)
    return strip(s)
end

readfile("day6.example")
```

```julia
split_lines(s::String)::Vector{String} = split(s, '\n')

split_lines(readfile("day6.example"))
```

```julia
split_parts(s::String)::Vector{String} = filter(p -> p != "", split(s, ' '))

split_parts.(split_lines(readfile("day6.example")))
```

```julia
function solve_part1(arr2d::Vector{Vector{String}})#::Int64
    res = []
    for i in 1:length(arr2d[1])
        is_add = last(arr2d)[i] == "+"
        r = is_add ? 0 : 1
        for j in 1:length(arr2d)-1
            n = parse(Int64, arr2d[j][i])
            if is_add
                r += n
            else
                r *= n
            end
        end
        push!(res, r)
    end
    sum(res)
end

solve_part1(split_parts.(split_lines(readfile("day6.example"))))
```
