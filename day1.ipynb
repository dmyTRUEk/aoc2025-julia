```julia
function readfile(filename::String)::String
    f = open(filename, "r")
    s = read(f, String)
    close(f)
    return strip(s)
end

readfile("day1.example")
```

```julia
struct InputLine
    is_left::Bool
    n::Int32
    
    function InputLine(s::String)
        is_left = (s[1] == 'L')
        n = parse(Int32, s[2:end])
        new(is_left, n)
    end
end
```

```julia
split_lines(s::String)::Array{String} = split(s, '\n')

split_lines("L10\nR22")
```

```julia
parse_all_input(s::String)::Array{InputLine} = InputLine.(split_lines(s))

parse_all_input("L10\nR22")
```

```julia
function solve_part1(input_lines::Array{InputLine})
    num::Int32 = 50
    res::Int32 = 0
    for input_line in input_lines
        num += (input_line.is_left ? -1 : 1) * input_line.n
        num = mod(num, 100)
        #println(num)
        if num == 0
            res += 1
        end
    end
    return res
end

solve_part1(parse_all_input(readfile("day1.example"))) # expected: 3
```

```julia
solve_part1(parse_all_input(readfile("day1.input")))
```
