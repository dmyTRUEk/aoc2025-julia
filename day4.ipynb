```julia
function readfile(filename::String)::String
    f = open(filename, "r")
    s = read(f, String)
    close(f)
    return strip(s)
end

readfile("day4.example")
```

```julia
function parse_map(s::String)::Matrix{Bool}
    lines = split(s, '\n')
    n = length(lines)
    m = fill(false, n, n)
    for (l, line) in enumerate(lines)
        for (c, char) in enumerate(line)
            m[l,c] = (char == '@')
        end
    end
    return m
end

parse_map(readfile("day4.example"))
```

```julia
function solve_part1(m::Matrix{Bool})::Int64
    N = size(m)[1]
    res = 0
    #m2 = fill(0, N, N)
    for l in 1:N
        for c in 1:N
            m[l,c] || continue
            n = 0
            n += get(m, (l-1, c-1), 0)
            n += get(m, (l-1, c), 0)
            n += get(m, (l-1, c+1), 0)
            n += get(m, (l, c-1), 0)
            n += get(m, (l, c+1), 0)
            n += get(m, (l+1, c-1), 0)
            n += get(m, (l+1, c), 0)
            n += get(m, (l+1, c+1), 0)
            if n < 4
                res += 1
            end
            #m2[l,c] = n
        end
    end
    return res
end

solve_part1(parse_map(readfile("day4.example")))
```

```julia
solve_part1(parse_map(readfile("day4.input")))
```

```julia
function solve_part2(m::Matrix{Bool})::Int64
    N = size(m)[1]
    res = 0
    while true
        m_prev = copy(m)
        for l in 1:N
            for c in 1:N
                m[l,c] || continue
                n = 0
                n += get(m, (l-1, c-1), 0)
                n += get(m, (l-1, c), 0)
                n += get(m, (l-1, c+1), 0)
                n += get(m, (l, c-1), 0)
                n += get(m, (l, c+1), 0)
                n += get(m, (l+1, c-1), 0)
                n += get(m, (l+1, c), 0)
                n += get(m, (l+1, c+1), 0)
                if n < 4
                    m[l,c] = false
                    res += 1
                end
                #m2[l,c] = n
            end
        end
        m == m_prev && break
    end
    return res
end

solve_part2(parse_map(readfile("day4.example")))
```
